```python
from src.kmeans import load_kmeans_data
from src.util import serialize_assignments
from k_means_constrained import KMeansConstrained
import pandas as pd

df = load_kmeans_data('./test.csv')

df.iloc[:, 0]
```

```python
n = 40
total = len(df)
epsilon = 5
min_size = total // n - epsilon
max_size = total // n + epsilon

data = df.iloc[:, 1:]

kmc = KMeansConstrained(n, min_size, max_size, random_state = 0)
kmc.fit_predict(data)
```

```python
items = df.iloc[:, 0]
labels = kmc.labels_

s = serialize_assignments(items, labels)
display(s)
```
